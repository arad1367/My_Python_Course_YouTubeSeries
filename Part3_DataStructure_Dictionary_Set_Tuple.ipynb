### 1. Dictionary

```python
# Dictionary example --> key:value pairs
my_dictionary = {"Car": ["Benz", "BMW", "Kia"],
                 "Colors": ["Black", "White"],
                 "Licence": True,
                 "Total Avg Prod": 1000}
my_dictionary

# access to keys
my_dictionary.keys()

# access to values
my_dictionary.values()

# Access to keys:values --> items() give us a list of all dictionary key:value
my_dictionary.items()

# Mutability
# print(f"previous version of dic: {my_dictionary}")
# my_dictionary["Price"] = 2000 # To define a new key:value
# print(f"New version of dic: {my_dictionary}")

my_dictionary["License"] = False
my_dictionary
```

```python
# Dictionary methods
print(my_dictionary)

# pop --> return a value that can be even store as a variable
# license_permission = my_dictionary.pop("License")
# print(f"Status of my license permission: {license_permission}")

# popitem --> to pop last key:value by default
# my_dictionary.popitem()
# my_dictionary

# update -->  update a key:value and even add new one if not exist
# my_dictionary.update({"Rate": 4.87})
# my_dictionary
my_dictionary.update({"License":True})
my_dictionary
```

### 2. Tuple

```python
# tuple example
my_tuple = (1, 2, 3, 4, 5)
print(type(my_tuple))

# tuple is immutable
try:
  my_tuple[0] = 1000
except TypeError as err:
  print(err)
  print("Tuple is immutable")

# slicing
my_tuple[::-1]
```

### Set

```python
# Set example --> Set return just unique values
my_set = {1, 1, 1, 5, 7, 9, 9}
my_set

# Sets method in python: https://www.w3schools.com/python/python_ref_set.asp
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set1 & set2 # intersection
set1 | set2 # Union
```
