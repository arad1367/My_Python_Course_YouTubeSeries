# Python Data Structures Cheat Sheet

| Type | Ordered | Mutable | Duplicates | Example |
| :--- | :---: | :---: | :---: | :--- |
| **List** | ✅ | ✅ | ✅ | `[1, 2, 3]` |
| **Tuple** | ✅ | ❌ | ✅ | `(1, 2, 3)` |
| **Set** | ❌ | ✅ | ❌ | `{1, 2, 3}` |
| **Dict** | ✅* | ✅ | Keys: ❌ | `{'a': 1}` |

---

### Detailed Breakdown

1. **List**
   - *Ordered, mutable, allows duplicates*
   - `my_list = [1, 2, 3]`

2. **Tuple**
   - *Ordered, immutable, allows duplicates*
   - `my_tuple = (1, 2, 3)`

3. **Set**
   - *Unordered, mutable, no duplicates*
   - `my_set = {1, 2, 3}`

4. **Dictionary (dict)**
   - *Key–value pairs, insertion-ordered (Python 3.7+)*
   - `my_dict = {"name": "Alice", "age": 25}`

```python
# List slicing
my_list[:3]
my_list[::-1] # [start:stop:step]
my_list[4:6:1]

# List mutability vs. string
my_list[0] = "Hungary"
my_list
```
